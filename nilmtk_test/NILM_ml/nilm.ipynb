```python
from utils import visualize_predictions, prepare_data, visualize_data
from metrics import evaluate_regression
from sklearn.linear_model import LinearRegression
```

```python
# 1 mains
# 2 fridge
# 3 electric oven
# 4 kettle
# 5 stove
path_main = "data/1.csv"
path_target = "data/3.csv"
window_size = 10
train_size = 0.7
```

```python
visualize_data(path_main)
```

```python
visualize_data(path_target)
```

```python
X, y = prepare_data(main_file=path_main, target_file=path_target, time_window=window_size)
```

```python
train_index_end = int(len(X)*train_size)
x_train = X[:train_index_end].to_numpy()
x_test = X[train_index_end:].to_numpy()
y_train = y[:train_index_end].to_numpy()
y_test = y[train_index_end:].to_numpy()
```

```python
linreg = LinearRegression()
linreg.fit(x_train, y_train)
y_pred = linreg.predict(x_test)
```

```python
evaluate_regression(y_pred, y_test)
```

```python
visualize_predictions(y_pred, y_test)
```
